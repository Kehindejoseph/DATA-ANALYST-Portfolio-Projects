```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv('/Users/k/Desktop/DATA ANALYTICS LEARNING/SQL + TABLEU PROJECT/RACE/ff_race_50.csv')
df.head()
```

```python
df2 = df.dropna(axis=1)
```

```python
df2.head()
```

```python
df2 ['fullName'] = df2['First'] + ' ' + df2['Last']
df2.head()
```

```python
df2['Time'] = pd.to_timedelta(df2['Time'])
```

```python
df2['Total_Minutes'] = df2['Time'].dt.total_seconds() / 60
```

```python
df2['Total_Minutes'] = df2['Total_Minutes'].round().astype(int)
```

```python
df2.head()
```

```python
df2.rename(columns={'Division': 'Gender'}, inplace=True)
```

```python
df2.head()
```

```python
def get_lat_long(city, state):
    # Combine city and state into an address
    address = f"{city}, {state}"
    try:
        # Initialize Nominatim object with the correct parameter name
        geolocator = Nominatim(user_agent="running", timeout=10)
        location = geolocator.geocode(address)
        if location:
            # Extract latitude and longitude from location object
            latitude = location.latitude
            longitude = location.longitude
            return latitude, longitude
        else:
            # Handle case where location is not found
            return None, None
    except Exception as e:
        # Handle exceptions
        print(f"Error: {e}")
        return None, None
```

```python
df2['latitude'], df2['longitude'] = zip(*df2.apply(lambda x: get_lat_long(x['City'], x['State']), axis=1))
```

```python
df2.head()
```

```python
# Apply the get_lat_long function to each row of the DataFrame
df2[['latitude', 'longitude']] = df2.apply(lambda x: pd.Series(get_lat_long(x['City'], x['State'])), axis=1)
```

```python
df2.head()
```

```python
df2['latlong'] = df2['latitude'].astype(str) + ', ' + df2['longitude'].astype(str)
```
